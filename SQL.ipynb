```python
import sqlite3 as sql
import pandas as pd
from google.colab import files
from sqlite3.dbapi2 import Cursor
```

```python
df = pd.read_csv("/content/tienda.csv")
df
```

```python
#se conecta a un database
conn =sql.connect('tienda.db')

#se pasa el df a sql
df.to_sql('oct19', conn, if_exists="replace", index = False)

conn.close()
print("Cree la base de datos llamada 'tienda'.\n Tambien cree una tabla llamada 'oct19'")
```
