```python
#Leer datos de base de datos
import pandas as pd
# Visualización y gráficos
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from google.colab import files
```

```python
#convertir el arcchivo en un DataFrame
Oct2019_df = pd.read_csv("/content/2019-Oct.csv", header = 0)
Oct2019_df
```

```python
#Obtener inforamción del DataFrame
Oct2019_df.info()
```

```python
#Observar que columnas tenemos en el DataFrame
Oct2019_df.columns
```

```python
#Crear una copia del DataFrame
Oct19_df = Oct2019_df[:600000].copy()
Oct19_df
```

```python
Oct19_df['event_time'] = pd.to_datetime(Oct19_df['event_time'])
```

```python
#Ver la información de como quedó la copia del DataFrame
Oct19_df.info()
```

```python
#Eliminar las columnas que no se van a utilizar
Oct19_df = Oct19_df.drop(["category_id", "user_session","user_id"], axis = 1)
Oct19_df
```

```python
#Observar las columnas que quedaron
Oct19_df.columns
```

```python
#Observar la cantidad de datos nulos
Oct19_df.isnull().sum()
```

```python
#Gráficar los datos nulos
sns.heatmap(Oct19_df.isnull(), cbar=False)
plt.show()
```

```python
#Eliminar las filas donde hay datos nulos
Oct19_df = Oct19_df.dropna(subset=["category_code", "brand"]) #dropna
```

```python
#Ver los datos nulos despues de la limpieza
Oct19_df.isnull().sum()
```

```python
#Gráficar los fatos nulos despues de la limpieza
sns.heatmap(Oct19_df.isnull(), cbar=False)
plt.show()
```

```python
#Observar las estadisticas del DataFrame
estadistica = Oct19_df.describe()
print(estadistica["price"])
```

```python
#Agrupar los valores de la columna para poderlos gráficar
evento = Oct19_df["event_type"].value_counts()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(evento, labels=evento.index, autopct='%1.1f%%')
plt.title("Distribución de Tipos de Eventos")
plt.show()
```

```python
marca = Oct19_df["brand"].value_counts()

top_n = 10
top_marcas = marca.head(top_n)

plt.figure(figsize=(6, 6))
plt.pie(top_marcas, labels=top_marcas.index, autopct='%1.1f%%')
plt.title("Distribución de marcas")
plt.show()
```

```python

categoria = Oct19_df["category_code"].value_counts()

# Seleccionar las 10 marcas más comunes (puedes ajustar este número)
top_n = 10
top_categoria = categoria.head(top_n)

plt.figure(figsize=(8, 6))

# Crear un gráfico de barras con las marcas más comunes
plt.bar(top_categoria.index, top_categoria, color= ["red", "grey", "violet", "yellow", "navy", "blue", "brown", "pink", "green", "orange"])

plt.xlabel("categoria")
plt.ylabel("Cantidad")
plt.title(f"Top {top_n} Categorias")
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mayor legibilidad

plt.show()
```

```python
Precio = Oct19_df["price"]
```

```python
# Crear el histograma
plt.hist(Precio, bins=5, color= 'purple', alpha=0.7)
plt.title('Histograma de los precios')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar el histograma
plt.show()
```
